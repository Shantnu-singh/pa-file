```python
# Encoding categorical data ( Normial--> states, color and categorical--> rating , class , review)
# Ordinal Encoding --> for input ordinal data
# apply seperatly on x_train and x_test

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# handling category with may categorical values
count = df['make'].value_counts()
threshold = 10
repl = count[count <= threshold].index
print(repl)

df['make'] = df['make'].replace(repl , "uncommon")
dummies = pd.get_dummies(df["make"] , dtype= np.int32)

df = pd.concat([df , dummies] , axis=1)

df.drop(columns= ['make'] , inplace= True)
```

```python
# train test split
X = df.drop(columns= ['price'])
y = df['price']

X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=.2)
```

```python
# better way to do this column tranformer 
from sklearn.compose import ColumnTransformer
one = OneHotEncoder(drop= "first" , sparse_output=False)
oe = OrdinalEncoder(categories=[['two' , 'four'] , ['tow' , 'two' , 'three' , 'four' , 'five' , 'six' , 'eight' , 'twelve']])

combined_df = pd.concat([X_test , X_train])

tranformer = ColumnTransformer(transformers=[
    ('tnf1' , one ,[ 'fuel_type' , 'body_style' , 'drive_wheels' , 'engine_location' , 'fuel_system' , 'engine_type'] ),
    ("tnf2" , oe , ['num_of_doors' , 'num_of_cylinders' ])

] , remainder="passthrough")
```

```python
combiend_encoded = tranformer.fit_transform(combined_df)

X_test = combiend_encoded[:len(X_test)]
X_train= combiend_encoded[len(X_test):]
```

```python
# Principle Component analysis

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()

X_train = scalar.fit_transform(X_train)
x_test = scalar.transform(X_test)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=None)

X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
```
