```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("automobile_data.csv")
df.sample(5)
```

```python
df= df.select_dtypes("number")
```

```python
df.isnull().sum()
df.duplicated().sum()
df.drop_duplicates(keep = "first" , inplace=True)
```

```python
# Standarrdization (feature scaling)
# Why
    # algo like knn (dis)  , logistice regression one factor domintates
# Standardization( Z score normlaisation) and Normilisation

# xi - xbar / verience -->create now column , mean == 0 , varience = 1
# this doesn't change curve disrtibtuin shape
# Can't handle outliers

# when to apply -- > K-means , KNN , PCA , ANN , Gradient Decent

```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop(columns= ["price"])
y = df['price']
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size= .2)
```

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
X_train_sc = pd.DataFrame(X_train_sc  , columns= X_train.columns)
X_test_sc = pd.DataFrame(X_test_sc , columns= X_test.columns)
```

```python
# Normlisation ( min- max scaling , mean normalisation , max )
# formula xi = xi-xmin / (xmax - xmin) --> 0 to 1 value range
# outerlier influence the data as it is also get scaled 

# mean normalisation xi = xi-mean / xmax - xmin ( mean centric ) -1 to 1 

# max abs scaling xi = xi/xmax --> useful for sparse data

# roboust scaling xi = xi-xmeadain / IQR  --> robust to outliers 
  
```
